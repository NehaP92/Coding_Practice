```python
import sqlite3
from sqlite3 import Error
import pandas as pd
```

```python
conn = sqlite3.connect('Mod_4_Blog_Database.db')
```

```python
df = pd.read_excel('Monthly_Data.xlsx')
df.to_sql('monthly_data',conn,if_exists='fail')
```

```python
cur=conn.cursor()
```

```python
def query_full(query):
    """
    Executes the sql query to output as pandas DataFrame
    ---------------------------------------
    Input:
    query (str): SQL query
    ---------------------------------------
    Output:
    pandas DataFrame
    """
    data = pd.DataFrame(cur.execute(query).fetchall())
    data.columns = [x[0] for x in cur.description]
    data.index=data['ID']
    data.drop(['ID','index'], axis=1,inplace = True)
    
    return data
    
```

```python
def query(query):
    """
    Executes the sql query to output as pandas DataFrame
    ---------------------------------------
    Input:
    query (str): SQL query
    ---------------------------------------
    Output:
    pandas DataFrame
    """
    data = pd.DataFrame(cur.execute(query).fetchall())
    data.columns = [x[0] for x in cur.description]
    
    return data
    
```

```python
energy=['Hydroelectric','Geothermal', 'Solar',
        'Wind','Wood', 'Waste', 'Biofuels']
df_energy=[]
for source in energy:
    df_energy.append(query(f"""SELECT ID, Month, {source} FROM monthly_data"""))
    
df_hyd,df_geo,df_solar,df_wind,df_wood,df_waste,df_biofuel = df_energy[0],df_energy[1],df_energy[2],df_energy[3],df_energy[4],df_energy[5],df_energy[6]

display(df_hyd)
display(df_geo)
display(df_solar)
display(df_wind)
display(df_wood)
display(df_waste)
display(df_biofuel)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = pd.read_csv("population.csv")
population
```

```python
pop_mod = population[population['Country Name'].isin(['Afghanistan','Arab World','Euro area'])]
pop_mod
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(x=pop_mod[pop_mod['Country Name']=="Afghanistan"]['Year'], 
           y=pop_mod[pop_mod['Country Name']=="Afghanistan"]['Value'], label="Afghanistan")
ax.scatter(x=pop_mod[pop_mod['Country Name']=="Arab World"]['Year'], 
           y=pop_mod[pop_mod['Country Name']=="Arab World"]['Value'],label="Arab World")
ax.scatter(x=pop_mod[pop_mod['Country Name']=="Euro area"]['Year'], 
           y=pop_mod[pop_mod['Country Name']=="Euro area"]['Value'],label="Euro Area")
plt.title('Population per Country')
plt.ylabel('Population')
plt.xlabel('Year')
plt.legend();
```

```python
sns.scatterplot(x='Year',y='Value',hue='Country Name', data=pop_mod,palette='mako');
plt.title('Population per Country')
plt.ylabel('Population')
plt.xlabel('Year')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(pop_mod[pop_mod['Country Name']=="Afghanistan"]['Year'], 
           pop_mod[pop_mod['Country Name']=="Afghanistan"]['Value'], label="Afghanistan")
ax.plot(pop_mod[pop_mod['Country Name']=="Arab World"]['Year'], 
           pop_mod[pop_mod['Country Name']=="Arab World"]['Value'],label="Arab World")
ax.plot(pop_mod[pop_mod['Country Name']=="Euro area"]['Year'], 
           pop_mod[pop_mod['Country Name']=="Euro area"]['Value'],label="Euro Area")
plt.title('Population per Country')
plt.ylabel('Population')
plt.xlabel('Year')
plt.legend();
```

```python
sns.lineplot(x='Year',y='Value',hue='Country Name', data=pop_mod,palette='mako');
plt.title('Population per Country')
plt.ylabel('Population')
plt.xlabel('Year')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.bar(x=pop_mod[pop_mod['Country Name']=="Afghanistan"]['Year'], 
           height=pop_mod[pop_mod['Country Name']=="Afghanistan"]['Value'], label="Afghanistan")
ax.bar(x=pop_mod[pop_mod['Country Name']=="Arab World"]['Year'], 
           height=pop_mod[pop_mod['Country Name']=="Arab World"]['Value'],label="Arab World")
ax.bar(x=pop_mod[pop_mod['Country Name']=="Euro area"]['Year'], 
           height=pop_mod[pop_mod['Country Name']=="Euro area"]['Value'],label="Euro Area")
plt.title('Population per Country')
plt.ylabel('Population')
plt.xlabel('Year')
plt.legend();
```

```python
pop_mod['decade']=pop_mod.apply(lambda x: int(x)+ int(x)%10)
pop_mod['decade']
```

```python
fig, ax = plt.subplots(figsize=(12,6))
plot = sns.barplot(x='Year',y='Value',hue='Country Name', data=pop_mod,palette='mako', ax=ax);
plt.title('Population per Country')
plt.ylabel('Population')
plt.xlabel('Year')
plot.set_xticklabels(plot.get_xticklabels(),rotation=90);
```

```python
plt.hist(pop_mod['Value'])
plt.title('Population Distribution')
plt.xlabel('Population');
```

```python
sns.distplot(pop_mod['Value'])
plt.title('Population Distribution')
plt.xlabel('Population');
```
